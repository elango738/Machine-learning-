```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
# Separate by species
df_Setosa = df[df['Species'] == 'Iris-setosa']
df_Versicolor = df[df['Species'] == 'Iris-versicolor']
df_Virginica = df[df['Species'] == 'Iris-virginica']
```

```python
# Scatter plot (with offsets so points don't overlap) univariant
plt.scatter(df_Setosa['SepalWidthCm'], np.zeros_like(df_Setosa['SepalWidthCm']),
            label='Setosa', color='red')
plt.scatter(df_Versicolor['SepalWidthCm'], np.ones_like(df_Versicolor['SepalWidthCm']),
            label='Versicolor', color='green')
plt.scatter(df_Virginica['SepalWidthCm'], np.full_like(df_Virginica['SepalWidthCm'], 2),
            label='Virginica', color='blue')

plt.xlabel('Sepal Width (cm)')
plt.yticks([0, 1, 2], ['Setosa', 'Versicolor', 'Virginica'])
plt.legend()
plt.show()
```

```python
# Bivariate scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(df_Setosa['SepalWidthCm'], df_Setosa['SepalLengthCm'], color='red', label='Setosa')
plt.scatter(df_Versicolor['SepalWidthCm'], df_Versicolor['SepalLengthCm'], color='green', label='Versicolor')
plt.scatter(df_Virginica['SepalWidthCm'], df_Virginica['SepalLengthCm'], color='blue', label='Virginica')

plt.xlabel('Sepal Width (cm)')
plt.ylabel('Sepal Length (cm)')
plt.title('Bivariate: Sepal Width vs Sepal Length')
plt.legend()
plt.show()
```

```python
sns.FacetGrid(df, hue="Species", height=5) \
   .map(plt.scatter, "PetalLengthCm", "SepalWidthCm") \
   .add_legend()

plt.show()
```
