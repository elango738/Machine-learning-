```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
```

```python
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_scaled)

pca_3d = PCA(n_components=3)
X_pca_3d = pca_3d.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
colors = ['red', 'green', 'blue']
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_pca_2d[y == i, 0], X_pca_2d[y == i, 1],
                color=color, label=target_name, s=60, alpha=0.8, edgecolors='k')
plt.title(" PCA - 2D Projection of Iris Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot(111, projection='3d')
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    ax.scatter(X_pca_3d[y == i, 0], X_pca_3d[y == i, 1], X_pca_3d[y == i, 2],
               color=color, label=target_name, s=60, alpha=0.8, edgecolors='k')
ax.set_title(" PCA - 3D Projection of Iris Dataset")
ax.set_xlabel("PC 1")
ax.set_ylabel("PC 2")
ax.set_zlabel("PC 3")
ax.legend()
plt.show()
```
